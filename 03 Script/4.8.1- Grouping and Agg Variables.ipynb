## Creating loyalty flag for existing customers using transform() and loc() functions. 

```python
# Creating "max_order" columns. 

ords_prods_merged['max_order']= ords_prods_merged.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
# Using loc()function to create loyalty flag

ords_prods_merged.loc[ords_prods_merged['max_order']> 40, 'loyalty_flag']= 'Loyal customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['max_order']<= 40) & (ords_prods_merged['max_order']>10),'loyalty_flag']= 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['max_order'] <= 10, 'loyalty_flag'] = 'New customer'
```

##Create a spending flag for each user based on the average price across all their orders using the following criteria. 

#If the mean of the prices pf products purchased by a user is lower than 10, then flag them as "Low spender."
#If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as "High spender."

```python
# Step 6: Creating a "mean_price" column

ords_prods_merged['average_price']= ords_prods_merged.groupby(['customer_id'])['prices'].transform(np.mean)
```

```python
#Using loc()function to create flag

ords_prods_merged.loc[ords_prods_merged['average_price']>= 10, 'spending_flag']= 'High spender'
```

```python
ords_prods_merged.loc[ords_prods_merged['average_price'] <10,'spending_flag'] = 'Low spender'
```

```python
#Using the value_counts()function to print frequency of the flag

ords_prods_merged['spending_flag'].value_counts(dropna = False)
```

##The result indicates that there are more low spender than hig spender.

##Creating an order frequency flag that marks the regularity of a user's ordering behavior according to the mediam in the the "days_since_prior_order" columnm, based on criteria;

#If the median of 'days_since_prior_order' is higher than 20, then the customer should be labeled a 'Non-frequent customer.'

#If the median is higher than 10 and lower than 20, then the customer should be labeled a 'Regualar customer.'

#If the meidan is lower than or equal to 10, then the cusotmer should be labeled a 'Frequent customer'

```python
#Step 7: Creating an 'order_frequency_flag'column.

ords_prods_merged['regularity']=ords_prods_merged.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Using .loc to create 'order_frequency_flag' column

ords_prods_merged.loc[ords_prods_merged['regularity']> 20,'order_frequency_flag'] = 'Non-frequent customer'
```

```python
ords_prods_merged.loc[(ords_prods_merged['regularity']>10) & (ords_prods_merged['regularity'] <=20),'order_frequency_flag']= 'Regular customer'
```

```python
ords_prods_merged.loc[ords_prods_merged['regularity']<=10,'order_frequency_flag']= 'Frequent customer'
```

```python
#Printing df using .value_counts()function

ords_prods_merged['order_frequency_flag'].value_counts(dropna = False)
```
