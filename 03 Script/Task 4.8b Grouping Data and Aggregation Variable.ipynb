# Creating loyalty flag for existing customers using transform() and loc() functions.

##Step 5: What is the spending habits of the three types of customer groups identified? ##Determine if the prices of products purchased by loyal customers differ from those purchased by regular customer.

We can see that Regular cutomers numbers are greater than both Loyal and New customer. 

```python
# Creating "max_order" columns. 

df['max_order']= df.groupby(['customer_id'])['order_number'].transform(np.max)
```

```python
df.loc[df['max_order']> 40, 'loyalty_flag'] = 'Loyal customer'
```

```python
df.loc[(df['max_order']<= 40) & (df['max_order']>10), 'loyalty_flag']= 'Regular customer'
```

```python
df.loc[df['max_order']<=10, 'loyalty_flag']= 'New customer'
```

```python
#Printing dataframe head()

df.head()
```

```python
#Using value_counts()function to view frequency of loyalty flag

df['loyalty_flag'].value_counts(dropna = False)
```

##Create a spending flag for each user based on the average price across all their orders using the following criteria.

#If the mean of the prices pf products purchased by a user is lower than 10, then flag them as "Low spender." #If the mean of the prices of products purchased by a user is higher than or equal to 10, then flag them as "High spender."

# Creating an "average_price" column

```python
df['average_price'] = df.groupby(['prices'])['customer_id'].transform(np.mean)
```

```python
#Using the loc()function to create average_price flags

df.loc[df['average_price']> 10, 'spending_flag']= 'High spender'
```

```python
df.loc[df['average_price']<= 10, 'spending_flag']= 'Low spender'
```

```python
#Printing "average_price" frequency using the value_counts()function

df['spending_flag'].value_counts(dropna = False)
```

```python
#Printing head of dataframe using head()function

df.head()
```

##Creating an order frequency flag that marks the regularity of a user's ordering behavior according to the mediam in the the "days_since_prior_order" columnm, based on criteria;

#If the median of 'days_since_prior_order' is higher than 20, then the customer should be labeled a 'Non-frequent customer.'

#If the median is higher than 10 and lower than 20, then the customer should be labeled a 'Regualar customer.'

#If the meidan is lower than or equal to 10, then the cusotmer should be labeled a 'Frequent customer'

```python
#Step 7: Create an 'order_frequency'column

df['regularity'] = df.groupby(['customer_id'])['days_since_prior_order'].transform(np.median)
```

```python
#Using loc()function to create 'order_frequency_flag' column

df.loc[df['regularity']>20, 'order_frequency_flag'] = 'Non_frequent customer'
```

```python
df.loc[(df['regularity']>10) & (df['regularity'] <=20), 'order_frequency_flag']= 'Regular customer'
```

```python
df.loc[df['regularity']<=10,'order_frequency_flag']= 'Frequent customer'
```

```python
#Printing order_frequency_flag using value_counts()function

df['order_frequency_flag'].value_counts(dropna = False)
```
